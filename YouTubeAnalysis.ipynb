# No of Channels in Each Category

```python
cdf = data.groupby("Category").size().reset_index(name="channel_count") \
    .sort_values("channel_count", ascending=False).head(20)

fig, ax = plt.subplots(figsize=(8,8))
_ = sns.barplot(x="channel_count", y="Category", data=cdf,
                            palette=sns.cubehelix_palette(n_colors=20, reverse=True), ax=ax)
_ = ax.set(xlabel="No. of Channels", ylabel="category")
```

# Which Category Has more Channels in Top 20 Subscribers

```python
a1 = data.sort_values("Subscribers (in M)",ascending=False).head(20)[['ChannelName','Views','Category']]
res1 = a1.groupby('Category').size().reset_index(name='Channel Count')
b1 = res1.sort_values("Channel Count",ascending=False)
fig, ax = plt.subplots(figsize=(8,8))
_ = sns.barplot(x="Channel Count", y="Category", data=b1,
                            palette=sns.cubehelix_palette(n_colors=20, reverse=True), ax=ax)
_ = ax.set(xlabel="No. of Channels", ylabel="category")
```

# Which Category has more Channels in Top 20 Views

```python
a2 = data.sort_values("Views",ascending=False).head(20)[['ChannelName','Views','Category']]
res2 = a2.groupby('Category').size().reset_index(name='Channel Count')
b2 = res2.sort_values("Channel Count",ascending=False)
fig, ax = plt.subplots(figsize=(8,8))
_ = sns.barplot(x="Channel Count", y="Category", data=b2,
                            palette=sns.cubehelix_palette(n_colors=20, reverse=True), ax=ax)
_ = ax.set(xlabel="No. of Channels", ylabel="category")
```

```python
b2 = pd.DataFrame(b2)
b2 = b2.reset_index(drop=True)
b2
```

```python
b1 = pd.DataFrame(b1)
b1 = b1.reset_index(drop=True)
b1
```

```python
cat = []
sums = []
for i in range(7):
    for j in range(6):
        if b1['Category'][i] == b2['Category'][j]:
            a = b1['Channel Count'][i]+ b2['Channel Count'][j]
            cat.append(b1['Category'][i])
            sums.append(a)
pop = pd.DataFrame({'Category': cat, 'Total': sums})
pop = pop.sort_values("Total",ascending=False).reset_index(drop=True)
```
